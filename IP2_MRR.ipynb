**Problem 19:** Write the code to get a list that contains each unique value in the weather column. 

```python
def unique_value(listx):
    list1 = []
    for x in listx:
        for word in x.split(','):
            if word not in list1:
                list1.append(word)
    for x in list1:
        return list1
```

```python
weather = unique_value(df2.weather)
```
